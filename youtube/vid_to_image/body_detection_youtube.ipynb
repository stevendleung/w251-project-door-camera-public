```python
import cv2
import imutils 
import os
import glob
```

```python
def max_label(name, folder):
    '''Look at a folder and check the files with pattern "name_###.jpg" to extract the
    largest label present.'''
    
    path_pattern = os.path.join(folder, name + "_*.jpg")
    existing_files = glob.glob(path_pattern)
    if not existing_files:
        biggest_label = 0
    else:
        existing_labels = map(lambda s: int(s.split('_')[-1].split('.')[0]), existing_files)
        biggest_label = max(existing_labels)
    return biggest_label
```

```python
def crop_ppl_from_images(input_image, filename_to_write, folder_to_write = 'cropped_ppl'):

    
    image = cv2.imread(input_image) 

    image = imutils.resize(image, 
                       width=min(500, image.shape[1])) 

    hog = cv2.HOGDescriptor()  
    hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector()) 

    (humans, _) = hog.detectMultiScale(image,  
                                     winStride=(2,2), padding=(32,32), scale=1.05)

    for (x, y, w, h) in humans: 
#     cv2.rectangle(image, (x, y),  
#                   (x + w, y + h),  
#                   (0, 0, 255), 2) 
        crop_img = image[y:y+h, x:x+w]

    try:
        file_number = str(max_label(filename_to_write, folder_to_write) + 1)
        file_to_write = filename_to_write + '_' + file_number
        cv2.imwrite(folder_to_write + '/' + file_to_write + '.jpg', crop_img)
    
    except:
        #print(crop_img)
        print('no human detected in img {}'.format(input_image))
    
    return
```

```python
directory = os.getcwd()

for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"): 
        crop_ppl_from_images(input_image = filename, filename_to_write = 'cropped', )
        
    else:
        continue
```
